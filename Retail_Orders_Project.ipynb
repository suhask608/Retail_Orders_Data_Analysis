```python
#renaming column by converting upper case letters to lower Case letters.
df.columns = df.columns.str.lower()
df.columns
```

```python
#renaming column by removing spaces with underscore 
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
#deriving new columns discount, sales_price & profit
df['discount']=df['list_price']*df['discount_percent']/100 
df['sales_price'] = df['list_price'] - df['discount']
df['profit'] = df['sales_price']-df['cost_price']
df
```

```python
#converting object datatpe to date formate for "order_date" column            
df['order_date']=pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df.dtypes
```

```python
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
```
